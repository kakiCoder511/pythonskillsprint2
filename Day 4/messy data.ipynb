```python
import pandas as pd
```

```python
#load in a new dataset
df = pd.read_csv("students_data_messy.csv")
df.head()
```

```python
df.info()
df.shape
```

```python
#droping rows if the value is missing
df_dropped = df.dropna()
df_dropped.head()
```

```python
df_dropped_col = df.dropna(subset = ["Age"])
df_dropped_col.head()
```

```python
#Filling in missing values
df.head(20)
df["Age_filled"] = df["Age"].fillna(0)
df.head(20)
```

```python
#fill missing valuses woth some relevant data not 0
df["Sore_Filled"] = df["Score"].fillna(df["Score"].mean())
df.head(20)
```

```python
#Filtering 

#older than 30
older_than_thirty = df[df["Age"]> 17]
older_than_thirty.head(20)
```

```python
#filter with multiple conditions
# get the age and score from students
high_score = df[(df["Age"] > 17) & (df["Score"] > 85)]
print(high_score.head(20))
```

```python
df_renamed = df.rename(columns = {"Age" : "Age_Years", "Score" : "Test_Score"})
df_renamed.head()
```
