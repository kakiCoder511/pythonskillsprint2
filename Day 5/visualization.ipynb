```python
sns.histplot(tips["tip"],kde=True, bins=30)
```

```python
#barplot
#useful when dealing with categorical data
sns.barplot(x="sex", y ="tip", data= tips, hue="sex", palette = "YlGnBu")
```

```python
sns.boxplot(data= tips, x="day", y = "tip", hue="sex", palette="YlGnBu")
```

```python
sns.stripplot(data= tips, x="day", y = "tip", hue="sex", palette="YlGnBu", dodge=True)
```

```python
#join plot
sns.jointplot(x= "tip",y ="total_bill", data= tips, kind="reg")
```

```python
#task - to explore both of the below plots using the seaboron
#pairplot
sns.pairplot(tips, x_vars=["total_bill"], y_vars=["tip"], kind="reg")
```

```python
#heatmap
import matplotlib.pyplot as plt

corr = tips.corr(numeric_only=True)   # correlation matrix
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```
