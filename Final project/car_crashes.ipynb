```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
crashes = sns.load_dataset("car_crashes")
crashes.head()
```

```python
#Top 10 States by Speeding Crashes
#What are the top 10 speeding state?
top10_speeding = crashes.nlargest(10, "speeding")
sns.barplot(
    x="speeding", 
    y="abbrev", 
    data=top10_speeding,
    hue="abbrev",
    palette="Reds_r",
)
plt.title("Top 10 States by Speeding-Related Crashes")
plt.xlabel("Speeding-related Crashes")
plt.ylabel("State")

```

```python
top10_alcohol = crashes.nlargest(10, "alcohol")
sns.barplot(data=top10_alcohol, x="alcohol", y="abbrev",  hue="abbrev", palette="Reds_r")
plt.title("Top 10 States by Alcohol-related Crashes")
plt.xlabel("Alcohol-related crashes")
plt.ylabel("State")
```

```python
# alcohol-related crashes increase?
#alcohol-related speedind
sns.scatterplot(data=crashes, x="alcohol", y="total", hue="speeding", palette="YlGnBu")
plt.title("Alcohol-related Crashes vs Total Crashes ")
```

```python
#distracted
crashes["distracted"] = crashes["total"] - crashes["not_distracted"]

sns.scatterplot(
    data=crashes, 
    x="distracted", 
    y="total", 
    hue="alcohol", 
    palette="YlGnBu"
)
plt.title("Distracted Crashes vs Total Crashes")
plt.xlabel("Distracted crashes")
plt.ylabel("Total crashes")
```

```python
#distracted driving does not strongly align with alcohol-related crashes
```

```python
#insurance premiums and insurance losses
sns.scatterplot(
    data=crashes,
    x="ins_premium", 
    y="ins_losses",
    hue="total", 
    palette="YlGnBu"
)
plt.title("Insurance Premium vs Insurance Losses")
plt.xlabel("Average Insurance Premium")
plt.ylabel("Average Insurance Losses")
```

```python
#This plot shows that insurance costs are influenced by multiple factors beyond crash frequency, 
#such as state regulations and healthcare costs."
```
